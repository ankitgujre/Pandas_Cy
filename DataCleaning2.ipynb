```python
import pandas as pd
import numpy as np
```

```python
# cleaning wrong format
my_data = pd.DataFrame({
    "Name": ["Alice","Bob","Charlie","David","Eva" , "Frank", "Grace", "Heide"],
    "Score": [85,np.nan,95,70,np.nan,88,95,88],
    "Age": [25,-1,23,22,21,25,np.nan,25],
    "Date" : ['2023-01-01','2023-02-15','not-a-date','2023-03-20','','2023-05-30','2023-05-30','2023-05-30']
})
```

```python
df = pd.read_csv('my_data.csv')
df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')
print(df)
# in this code wrong format of date show as NaT (not a time)
```

```python
# removing column with wrong age format

df  = pd.read_csv('my_data.csv')
df = df[df['Age'] > 0 ]
df
```

```python
# remove duplicate rows
df = pd.read_csv('my_data.csv')
df.drop_duplicates(inplace = True)
df
```

```python
# working with columns
import pandas as pd
```

```python
data = {
    'Book':['1984','To Kill a mockingbird', 'The great gatspby', 'moby dick'],
    'Author':['George Orwell', 'Harper Lee', 'F.scott', 'Herman Melville'],
    'Genre':['Dystoppian','Classic','Classic','Adventure' ],
    'Price':[250,300,200,350]
}

df = pd.DataFrame(data)
df
```

```python
# selecting column
df[['Book','Price']]
```

```python
# adding column
publication_year = [1949,1960,1929,1851]
df['year'] = publication_year
df
```

```python
# deleting column
df.drop(['Genre'], axis = 1, inplace = True)
display(df)
```

```python
# selecting rows
row = df.loc[df['Book'] == 'moby dick']
row
```

```python
# Selecting rows
rows = df.loc[df['Book'].isin(['moby dick', 'The Great gatspy'])]
rows
```

```python
# Adding Rows
new_book = pd.DataFrame({
    'Book': ['Pride and Prejudice'],
    'Author': ['Jane Austen'],
    'Price' : [200],
    'Year': [1813]
})

df = pd.concat([df,new_book], ignore_index = True)
df
```
