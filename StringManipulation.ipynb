```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Employee':['Rajesh', 'Shubham', 'Ashish', 'Neha', 'Mehak', np.nan, 'Prashant'],
    'Department': ['HR','Finance', 'IT', 'HR', 'Marketing', 'IT','Finance']
    
}
df = pd.DataFrame(data)
df
```

```python
# changing column data type to string
df = df.astype('string')
print(df.dtypes)
```

```python
# convert to lowercase
print(df['Employee'].str.lower())
```

```python
# convert to upper case
print(df[['Employee', 'Department'].str.upper()])
```

```python
# Remove leading and trailing spaces
df['Employee'] = df['Employee'].str.strip()
print(df['Employee'])
```

```python
# splitting the string
df['Split_Employee'] = df['Employee'].str.split('a')
print(df[['Split_Employee','Employee']])
```

```python
# calculating string length
print(df['Employee'].str.len())
```

```python
# concatinate string
print(df['Department'].str.cat(set = '; '))
```

```python
# how many types of encoding
# one hot encoding
print(df['Department'].str.get_dummies())
```

```python
# checks starts with
print(df['Employee'].str.startswith('A'))
```

```python
# checks ends with
print(df['Employee'].str.endswith('a'))
```

```python
# replace substrings
print(df['Employee'].str.replace('Ashish', 'Om prakash'))
```

```python
# repeat strings
print(df['Employee'].str.repeat(2))
```

```python
# count occurence
print(df['Employee'].str.count('a'))
```

```python
# find substring position

print(df['Employee'].str.find('a'))
```

```python
# find all ocuurence
print(df['Employee'].str.findall('a'))
```

```python
print(df['Employee'].str.islower())  # check for lowercase
```

```python
print(df['Employee'].str.isupper())  # check for uppercase
```
