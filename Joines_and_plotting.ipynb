```python
import pandas as pd
df1 = pd.DataFrame({
    'user_id':[101,102,103,104],
    'name':['Ashish','Vikash','Rajesh','Prashant']
})
df2 = pd.DataFrame({
    'user_id':[101,102,105,106],
    'City':['Delhi','Mumbai','Chennai','Bhopal']
})
display(df1)
display(df2)
```

```python
# Inner join : return only the rows with key present in both 'dfs'
result = pd.merge(df1,df2, on = 'user_id', how = 'inner')
print(result)
```

```python
# Left join: return left matching row + left table rows
result = pd.merge(df1,df2, on = 'user_id', how = 'left')
print(result)
```

```python
# Right join: return Right matching row + left table rows
result = pd.merge(df1,df2, on = 'user_id', how = 'right')
print(result)
```

```python
# Full outer join: all rows from both df
result = pd.merge(df1,df2, on = 'user_id', how = 'outer')
print(result)
```

```python
# index join : join dataframes based on index
result = pd.merge(df1, df2, left_index = True, right_index = True, how = 'inner')
result
```

## plotting

```python
import pandas as pd
import numpy as np
df_sale = pd.DataFrame({
    'Month':['jan','feb','mar','apr','may'],
    'Product_A':[120,150,170,130,160],
    'Product_B':[100,130,140,120,150]
}).set_index('Month')
df_random = pd.DataFrame({
    'X': np.random.normal(50,15,200),
    'Y':np.random.normal(30,10,200)
})
display(df_sale)
display(df_random)
```

```python
# line plot
df_sale.plot()
```

```python
df_sale.plot.area(alpha = 0.5)
```

```python
# bar plot
df_sale.plot.bar()
```

```python
# scatter plot
df_random.plot.scatter(x = 'X', y = 'Y', color = 'purple')
```

```python
# histogram: frequency distribution
df_random['X'].plot.hist(bins = 20, color = 'blue')
```

```python
# boxplot sumarize distribution, showing median, quartile and outlier
df_sale.plot.box()
```

```python
# hexbin plot usefull for large data set with overlapping point
df_random.plot.hexbin(x = 'X', y = 'Y', gridsize = 25, cmap = 'Blues')
```

```python
# KDE kernal density estimation show the probability density of a variable
df_random['X'].plot.kde()
```
