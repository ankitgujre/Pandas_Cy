```python
import pandas as pd
import numpy as np
```

```python
data = {
    'math':[85,92,np.nan, 78],
    'science':[88,np.nan,91,84],
    'English':[np.nan,75,89,95]
}
df = pd.DataFrame(data)
missing = df.isnull()
display(missing)
```

```python
# filtering all rows with missing values
students = pd.DataFrame(
    {
        'Name':['Ashish','Rakhi','Dhun','Nitin'],
        'Gender':['M',None,"F",None]
    }
)
missing_gender = students[students['Gender'].isnull()]
print(missing_gender)


```

```python
# using isna
# detecting missing values using isna()

data = {
    'student':['Ajay','Ravi',np.nan,'Suraj'],
    'Score':[90,np.nan,85,88]
}
df = pd.DataFrame(data)
print(df.isna())
```

```python
# filling missing value using fillna()
df_filled = df.fillna(0)
print(df_filled)
```

```python
# using forward  fill (uss previous values)
df_fffill = df.fillna(method = 'ffill')
print(df_fffill)
df_fffill
```

```python
# Backward fill 
df_bfill = df.fillna(method = 'bfill')
print(df_bfill)
df_bfill
```

```python
# Specific value Fill

data = {
    'student':['Ajay','Ravi',np.nan,'Suraj'],
    'Score':[90,np.nan,85,88]
}
df = pd.DataFrame(data)
```

```python
fillspecific = df['Score'].fillna('unknown')
print(fillspecific)
```

```python
# Replace value with replace() function
df_replace = df.replace(to_replace = np.nan, value = -1)
print(df_replace)
```

```python
# Using replace interpolate
data = {
    'A': [10,5,8,None,3],
    "B":[None, 7,15,9,None],
    "C":[12,11,None,6,4]
}
df = pd.DataFrame(data)
```

```python
df_interpolate = df.interpolate(method = 'linear', limit_direction = 'forward')
print(df_interpolate)
```
