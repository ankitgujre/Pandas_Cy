```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'Employee':['Ashish','Prashant','Rajesh','Vikash'],
    'Age':[28,34,29,40],
    'Department':['HR',"Engineering",'Finance', 'Marketing']
})
df2 = pd.DataFrame({
    'Employee':['Ashish','Prashant','Rajesh','Vikash'],
    'Salary':[70000,90000,60000,80000]
})
merged_df = pd.merge(df1,df2, on = 'Employee')
merged_df
```

```python
# Applying Custom Function
def high_income(salary):
    return 'High' if salary > 75000 else 'Low'

merged_df['IncomeLevel'] = merged_df['Salary'].apply(high_income)
merged_df
```

```python
# Lambda Function
max_age = merged_df['Age'].max()
merged_df['AgeNormalized'] = merged_df['Age'].apply(lambda x:x/max_age).round(2)
merged_df
```

```python
# visualizing
merged_df.plot(x = 'Employee', y = 'Salary', kind = 'scatter')
merged_df
```

```python
# Histogram
merged_df['Age'].plot.hist()
merged_df
```

## stats with pandas



```python
data = {
    'Height':[165,180,175,160],
    'Weight':[65,85,70,55],
    'Age':[28,34,29,22],
    'Score':[88,92,85,90]
}
df = pd.DataFrame(data)
df
```

```python
# min max scaling
# Scale calues between 0 to 1

min_max_scale = df.copy()
for col in min_max_scale.columns:
    min_val = min_max_scale[col].min()
    max_val = min_max_scale[col].max()
    min_max_scale[col] = (min_max_scale[col] - min_val)/(max_val - min_val)

min_max_scale
```
