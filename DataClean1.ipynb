```python
# data cleaning
my_data = pd.DataFrame({
    "Name": ["Alice","Bob","Charlie","David","Eva" , "Frank", "Grace", "Heide"],
    "Score": [85,np.nan,95,70,np.nan,88,95,88],
    "Age": [25,-1,23,22,21,25,np.nan,25],
    "Date" : ['2023-01-01','2023-02-15','not-a-date','2023-03-20','','2023-05-30','2023-05-30','2023-05-30']
})
```

```python
df = pd.DataFrame(my_data)
df.to_csv("my_data.csv", index = False)
```

```python

display(df)
```

```python
# removing Rows with empy cells

cleaned_df = df.dropna()
print(cleaned_df.to_string())
print(type(cleaned_df))
print(type(cleaned_df.to_string()))

```

```python
# replace nan values
df.fillna(999, inplace = True)
print(df)
```

```python
df = pd.read_csv('my_data.csv')
df.fillna({"Score": 999}, inplace = True)
display(df)
```

```python
# replacing column with mean value
```

```python
df = pd.read_csv('my_data.csv')
mean_value = df["Score"].mean()
df.fillna({"Score": mean_value.round(2)}, inplace = True)
display(df)
```

```python
# replacing column with median value
df = pd.read_csv('my_data.csv')
median_value = df["Score"].median()
df.fillna({"Score": median_value}, inplace = True)
display(df)
```

```python
import pandas as pd

df = pd.read_csv('my_data.csv')

# Calculate mode (returns a Series, so take the first value)
mode_value = df["Score"].mode()[0]

# Replace NaN values with mode
df.fillna({"Score": mode_value}, inplace=True)

display(df)
```
