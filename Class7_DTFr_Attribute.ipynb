```python
import numpy as np
import pandas as pd
```

```python
k = [90,20,3,4,90,23]
dt = pd.DataFrame(k)
print(dt)
dt
```

```python
k = [(19,20,21),(10,20,30)]
dt = pd.DataFrame(k, columns = ('m1','m2','m3')) # atribute/property
print(dt)
```

```python
w = [['sun','mon','tue'],['jan', 'feb','mar']]
dt = pd.DataFrame(w)
print(dt)
```

```python
arr = np.array(w)
arr
```

```python
tr = arr.T
tr
```

```python
dt = pd.DataFrame(tr, columns = ['week','month'])
print(dt)
```

```python
w = {'Weekdays':['sun','mon','tue'],'Months':['jan', 'feb','mar']}
dt = pd.DataFrame(w)
dt
```

```python
w = np.array([['sun','mon','tue'],['jan', 'feb','mar']]).reshape(3,2)
dt = pd.DataFrame(w)
dt
```
