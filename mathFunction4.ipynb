```python
import numpy as np
import pandas as pd
```

```python
# maths functions

arr = np.array([9,10,3,2,4,5])
sr = pd.Series(arr)
print(sr)
```

```python
print(sr.count())  # how many elements count
```

```python
print(sr.product())
```

```python
print(sr.sum())
```

```python
print(sr.mean())
```

```python
print(sr.mode())
```

```python
print(sr.min())
```

```python
print(sr.max())
```

```python
print(sr.std())
```

```python
print(sr.var())
```

```python
print(sr.describe())
```

```python
a = [9,10]
s = pd.Series(a)
print(s)
```

```python
print(s.var())
```

```python
# fancy indexing

k  = [30,12,3,2,4,56,7]
d = pd.Series(k)
d[5] = 65
print(d)
d[1::2] = 100
print(d)
```

```python
d[[1,3,6]] = 100
print(d)
```

```python
# null variations
j = [90,None,78,30,None,23]
g = pd.Series(j)
print(g)
```

```python
print(g.isnull())
```

```python
print(g.hasnans)
```

```python
print(g.notnull)
print(g.notnull())
```

```python
# unique and nunique

b = [12,3,5,3,6,7,12,90,8]
f = pd.Series(b)
print(f)
```

```python
print(f.unique())
```

```python
print(f.nunique())
```
